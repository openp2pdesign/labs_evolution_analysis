```python
import os
from makerlabs import fablabs_io
from makerlabs import hackerspaces_org
from makerlabs import makerspaces_make_co
from makerlabs import repaircafe_org
```

```python
orstoken = ""
with open('orstoken.txt', 'r') as f:
    orstoken = f.readline()
    orstoken = orstoken.rstrip()
```

```python
open_cage_api_key=orstoken
```

```python
diybios = diybio_org.get_labs(format="pandas", open_cage_api_key=open_cage_api_key)
```

```python
fablabs = fablabs_io.get_labs(format="pandas", open_cage_api_key=open_cage_api_key)
```

```python
hackerspaces = hackerspaces_org.get_labs(format="pandas", open_cage_api_key=open_cage_api_key)
```

```python
makerspaces = makerspaces_make_co.get_labs(format="pandas", open_cage_api_key=open_cage_api_key)
```

```python
repaircafes = repaircafe_org.get_labs(format="pandas", open_cage_api_key=open_cage_api_key)
```

```python
fablabs.to_csv("data/fablabs.csv")
hackerspaces.to_csv("data/hackerspaces.csv")
makerspaces.to_csv("data/makerspaces.csv")
repaircafes.to_csv("data/repaircafes.csv")
```
